```python
import pandas as pd
import numpy as np
```

```python
df=pd.read_csv("final_data.csv")
```

```python
df.head()
```

Checking whether there are null values in the dataset or not.

```python
df.isnull().sum()
```

Dropping the columns of object type. they have no contribution in training model.



```python
df=df.drop(columns=['player', 'team', 'name','position'])
df
```

```python
df.boxplot(column='current_value')
```

```python
X = df.drop(columns=["current_value"])
y=df['current_value']
```

```python
from sklearn.model_selection import train_test_split

X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.2,random_state=11)
```

Feature Scaling


```python
from sklearn.preprocessing import StandardScaler
sc=StandardScaler()
```

```python
X_train.iloc[:,:]=sc.fit_transform(X_train.iloc[:,:])
```

```python
X_train
```

```python
X_test.iloc[:,:]=sc.transform(X_test.iloc[:,:])
```

```python
X_test
```

```python
from sklearn.tree import DecisionTreeRegressor
dtc=DecisionTreeRegressor()
```

```python
dtc.fit(X_train,y_train)
```

```python
y_pred=dtc.predict(X_test)
```

```python
from sklearn.metrics import r2_score
```

```python
score=r2_score(y_test,y_pred)
score
```

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(14,7))
plt.plot(y_test,'lightgreen',label='Original value')
plt.plot(y_pred,'blue',label='Predicted value')
plt.legend()
plt.show()
```

Predicting current price for the new player.
Example:

```python
new_player = [180, 22, 34, 0, 0.033507074, 0.335070737, 0, 0, 0, 0, 2686, 175, 28, 1, 12000000, 2, 1]
new_player_array = np.array(new_player)
#Reshape the array to a 2D array
new_player_2d_array = new_player_array.reshape(1, -1)

prediction = dtc.predict(new_player_2d_array)
prediction
```
